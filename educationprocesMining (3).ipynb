```python
Good = df.loc[(df['TOTAL\n(100 points)'] >= 70.0)]
print(len(Good))

Fair = df.loc[(df['TOTAL\n(100 points)'] >= 51.0) & (df['TOTAL\n(100 points)'] <= 69.0)]
print(len(Fair))

Weak = df.loc[(df['TOTAL\n(100 points)'] <= 50.0)]
print(len(Weak))


# Assigning labels to the categories
df.loc[df['TOTAL\n(100 points)'] >= 70.0, 'label'] = 0
df.loc[(df['TOTAL\n(100 points)'] >= 51.0) & (df['TOTAL\n(100 points)'] <= 69.0), 'label'] = 1
df.loc[df['TOTAL\n(100 points)'] <= 50.0, 'label'] = 2

```

```python
df.drop(columns=[
    'Student ID',
    'ES 1.1 \n(2 points)',
    'ES 1.2 \n(3 points)',
    'ES 2.1\n(2 points)',
    'ES 2.2\n(3 points)',
    'ES 3.1\n(1 points)',
    'ES 3.2\n(2 points)',
    'ES 3.3\n(2 points)',
    'ES 3.4\n(2 points)',
    'ES 3.5\n(3 points)',
    'ES 4.1\n(15 points)',
    'ES 4.2\n(10 points)',
    'ES 5.1\n(2 points)',
    'ES 5.2\n(10 points)',
    'ES 5.3\n(3 points)',
    'ES 6.1\n(25 points)',
    'ES 6.2\n(15 points)'

],inplace=True)
```

```python
x = df.drop(['label'],axis=1).values
y = df['label'].values
```

```python
df.head()
```

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
##Feature Scaling
# Standardization
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x_train)
x_train_sc=sc.transform(x_train)
x_test_sc=sc.transform(x_test)
# Normaization
from sklearn.preprocessing import MinMaxScaler
mms= MinMaxScaler()
mms.fit(x_train)
x_train_mms=mms.transform(x_train)
x_test_mms=mms.transform(x_test)
```

```python

from sklearn.linear_model import Perceptron
p=Perceptron()

p.fit(x_train_sc,y_train)

train_pred=p.predict(x_train_sc)
test_pred=p.predict(x_test_sc)
```

```python
from sklearn.metrics import classification_report
cr = classification_report(y_test,test_pred)
print(cr)
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
cm = metrics.confusion_matrix(y_test,test_pred)
print(cm)
```

```python

sns.heatmap(cm,annot=True,fmt=".3f",linewidths=.5,square=True,cmap='Blues_r');
plt.xlabel("Predicted Label");
plt.ylabel("Actual Label");
plt.title('Confusion Matrix Heatmap')
```

```python

train_pred = l.predict(x_train)
test_pred = l.predict(x_test)
from sklearn.metrics import accuracy_score
train_accur = accuracy_score(y_train,train_pred)
test_accur = accuracy_score(y_test,test_pred)
```

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score, f1_score

permodel = Perceptron()
permodel.fit(x_train, y_train)
y_pred = permodel.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# F1 score
f1 = f1_score(y_test, y_pred, average='weighted')
print("F1 Score:", f1)
```

```python
# modelling

from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

# evaluating the model
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
print("RMSE Score :", np.sqrt(mse))

# r2 score
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print("R2 Score :", r2)
```

```python
# modelling

from sklearn.linear_model import LinearRegression

lmodel = LinearRegression()
lmodel.fit(x_train, y_train)

y_pred =lmodel.predict(x_test)

# evaluating the model
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
print("RMSE Score :", np.sqrt(mse))

# r2 score
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print("R2 Score :", r2)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import r2_score, mean_absolute_error

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
adamodel = AdaBoostClassifier(n_estimators=100, learning_rate=1)

adamodel.fit(x_train, y_train)
y_pred = adamodel.predict(x_test)

print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))

```
